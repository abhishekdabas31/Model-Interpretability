# Import Libraries

```python
import pandas as pd  #for manipulating data
import numpy as np  #for manipulating data
import sklearn  #for building models
import xgboost as xgb  #for building models
import sklearn.ensemble  #for building models
from sklearn.model_selection import train_test_split  #for creating a hold-out sample
from sklearn.metrics import mean_absolute_error # mean error
import time  #some of the routines take a while, so we monitor the time
import os  #needed to use Environment Variables in Domino
import matplotlib.pyplot as plt  #for custom graphs at the end
import seaborn as sns  #for custom graphs at the end
```

# Importing Data

```python
# Boston Housing Data
X,y = shap.datasets.boston()
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_display, y_display = shap.datasets.boston(display=True)
select = range(20)
features = X_test.iloc[select]
features_display = X_display.loc[features.index]
```

# Build the Models

1. [XGBoost](https://www.kaggle.com/alexisbcook/xgboost)

```python
xgb_model = xgb.train({'objective':'reg:squarederror'}, xgb.DMatrix(X_train, label=y_train))
```

```python
# make predictions
xgbmodel= xgb.XGBRegressor(n_estimators=500)
xgbmodel.fit(X_train,y_train)
predictions = xgbmodel.predict(X_test)

# calculatating error
print("Mean Absolute Error : " + str(mean_absolute_error(predictions, y_test)))
```

2. Random Forest regressor

```python
rf = sklearn.ensemble.RandomForestRegressor()
rf.fit(X_train, y_train)
```

```python
predictions_rf = rf.predict(X_test)

# calculatating error
print("Mean Absolute Error : " + str(mean_absolute_error(predictions_rf, y_test)))
```

3. K Nearest Neighbours

```python
knn = sklearn.neighbors.KNeighborsRegressor()
knn.fit(X_train, y_train)
```

```python
predictions_knn = knn.predict(X_test)

# calculatating error
print("Mean Absolute Error : " + str(mean_absolute_error(predictions_knn, y_test)))
```

### XGBoost seems the most accurate Model

## [SHAP Explainers](https://www.kaggle.com/dansbecker/shap-values)

```python
explainerXGB = shap.TreeExplainer(xgb_model)
shap_values_XGB_test = explainerXGB.shap_values(X_test)
shap_values_XGB_train = explainerXGB.shap_values(X_train)
```

## Explaining an instance (Local Interpretability)

```python
# j = np.random.randint(0, X_test.shape[0])

# optional, set j manually
j = 0

# initialize js
shap.initjs()
```

# SHAP force plot


```python
shap.force_plot(explainerXGB.expected_value, shap_values_XGB_test[j], X_test.iloc[[j]])
```

1. In the above figure, the Red values are the ones that are driving the prediction to increase, whereas the blue values are driving the prediction Lower.
2. The base value is the average prediciton of the Model, i.e "22.92" in this case

## Decision Plots

```python
shap.decision_plot(explainerXGB.expected_value, shap_values_XGB_test,features_display)
```

1. X- axis shows the predicted values, Y-axis shows the Model's Features (in descending order)
2. At the bottom is the expected value (from the imput parameter "explainerXGB.expected_value"), and at the top is the predicted value
3. Each observation is represented in a defferent color from the color spectrum. 
4. It show how Shap values are added to the models base value which is leading to the prediciton.

```python
# shap.decision_plot(explainerXGB.expected_value , shap_values_XGB_test , features_display , link='logit')
# shap.decision_plot(expected_value, shap_values, features_display, )
value= 1
r = shap.decision_plot(explainerXGB.expected_value , shap_values_XGB_test[value] , features_display.iloc[value] ,
                       return_objects=True, feature_order=r.feature_idx, xlim=r.xlim)
```

1. All values for that particular prediction are visible in this case, making it more viable to understand

### Change the base value

```python
p = 0.6  # Probability 0.4
new_base_value = np.log(p / (1 - p))  # the logit function
shap.decision_plot(explainerXGB.expected_value , shap_values_XGB_test,
                   features_display, link='logit', new_base_value=new_base_value)
```

1. Sometimes it might be case, where we want to change or shift the base value for making it more visually intuitive

## [SHAP Dependence Plot:](https://slundberg.github.io/shap/notebooks/plots/dependence_plot.html)
A dependence plot is a scatter plot that shows the effect a single feature has on the predictions made by the model.
It is similar to partial dependence plots, with some extra information

```python
# The first Parameter is the index of the Feature you want to see
# 2nd argument is the matrix of Shap values of all features
# 3rd argument is the data matrix
shap.dependence_plot(6, shap_values_XGB_train, X_train)
```

1. Everypoint is a sigle prediction matrix
2. X-axis hows the value of the feature and the Y-axis shows the SHAP values of the feature
3. The colored line shows the vertical coloring pattern of another feature, which shows that the 2 features have an interaction effect
4. The plot shows that with age the prediction almost remains the same, but the deviation in the start as well as the end, shows that - if age is less, and for same low value of LSTAT , the predicitons are lower. As age increases above 80 and same low value of LSTAT, the prediction may increase(higher chances) or decrease. But if with age and the LSTAT increases simultaneously the predictions do not vary much.

```python
# we can turn off interaction coloring
shap.dependence_plot("LSTAT", shap_values_XGB_train, X_train , interaction_index=None)
```

1. As the LSTAT value increases, it tends to affect the prediciton negatively (or reduces the prediction)
